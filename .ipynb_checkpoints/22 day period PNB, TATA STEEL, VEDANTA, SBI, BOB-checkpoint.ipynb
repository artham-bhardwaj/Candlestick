```python
#SBI

import pandas as pd
import plotly.graph_objs as go
# Read the CSV file into a DataFrame
df_sbi = pd.read_csv('SBIN.NS.csv')
df_sbi_close=df_sbi['Close']
df_sbi_volume=df_sbi['Volume']
```

```python
group_sbi_close=df_sbi_close.groupby(df_sbi_close.index//22).mean()
group_sbi_volume=df_sbi_volume.groupby(df_sbi_volume.index//22).mean()
```

```python
df_sbi_close.shape
```

```python
df_sbi_volume.shape
```

```python
group_sbi_close.shape
```

```python
group_sbi_close.shape
```

```python
candlestick_sbi=go.Candlestick(
open=group_sbi_close,
    close=group_sbi_close,
    high=group_sbi_close,
    low=group_sbi_close,
    increasing_line_color='green',
    decreasing_line_color='red',
    name='SBI stock price'
)
```

```python
#creating layout
layout_sbi=go.Layout(title='SBI STOCK PRICES',
                 yaxis_title='CLOSE'
                )
```

```python
#creating figure
fig_sbi=go.Figure(
data=[candlestick_sbi],
    layout=layout_sbi
)
```

```python
#create a volume trace
volume_sbi=go.Bar(x=group_sbi_volume.index,
             y=group_sbi_volume,
             name='Volume',
             marker_color='blue')
```

```python
#Add the volume trace if volume=True
volume_bool=True
if volume_bool:
    fig_sbi.add_trace(volume_sbi)
```

```python
fig_sbi.show()
```
