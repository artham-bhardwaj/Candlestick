```python
import pandas as pd
```

```python
df=pd.read_csv('adanicsv.csv')
```

```python
#changing index to date from integers
df=df.set_index(pd.DatetimeIndex(df['Date']))
```

```python
import plotly.graph_objects as go
```

```python
#creating an interactive candlestick chart
figure=go.Figure(
    data=[
        
        go.Candlestick(
        x=df.index,  #setting index as date values
        low=df['Low'],
        high=df['High'],
        close=df['Close'],
            open=df['Open'],
            increasing_line_color='green',
            decreasing_line_color='red'
        )
    ]
)
```

```python
#removing slider
figure.update_layout(xaxis_rangeslider_visible=False)

#giving titles
figure.update_layout(
title='Adani Stock Prices',
    xaxis_title='Date',
    yaxis_title='Stock prices'
)
```

```python
#adding volume bar
figure.add_trace(go.Bar(
       x=df['Date'],
       y=df['Volume'],
    marker_color='rgb(158,202,225)',
    name='Volume'
))
```

```python
figure.show()
```
