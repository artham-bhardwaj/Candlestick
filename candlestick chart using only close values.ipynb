```python
import pandas as pd
```

```python
import plotly.graph_objs as go
```

```python
#loading sample data in data frame
df=pd.read_csv("adanicsv.csv")
```

```python
#selecting only date for index , close values and volume values for volume bar
df=df[['Date','Close','Volume']]
```

```python
#df.set_index(df['Date'])  #to set index to dates rather than integers

#converting dates to standart datetime format
#df['Date']=pd.to_datetime(df['Date'])  #not working
```

```python
#creating candlestick trace
candlestick=go.Candlestick(
x=df['Date'],
    open=df['Close'],
    close=df['Close'],
    high=df['Close'],
    low=df['Close'],
    increasing_line_color='green',
    decreasing_line_color='red',
    name="ADANI STOCK PRICE"
)
```

```python
#creating layout : it sets the titles and axis titles
layout=go.Layout(title='ADANI STOCK PRICES',
xaxis_title='DATE',
yaxis_title='PRICE'
)
```

```python
#creating a volume bar
volume=go.Bar(x=df['Date'],
             y=df['Volume'],
              marker_color='black',
              name='Volume'
             )
```

```python
#creating figure : combines candlestick and voume traces and the layout
fig=go.Figure(
   data=[candlestick,volume],
    layout=layout
)
```

```python
fig.show()
```
